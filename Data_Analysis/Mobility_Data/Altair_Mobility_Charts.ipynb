```python
import pandas as pd
from pprintpp import pprint as pp
import numpy as np 
import altair as alt
from vega_datasets import data
import random
import glob
```

```python
mobility = pd.read_csv("mobility_data_for_altair.csv")
```

```python
mobility
```

```python
mobility['Date'] = pd.to_datetime(mobility.Date)
```

```python
mobility['Difference'] =  mobility['Difference'] / 100
```

```python
mobility['Change compared to baseline'] = mobility['Difference'] - 1
```

```python
nyc = mobility[mobility['Area'] == "New York City"]
mn = mobility[mobility['Area'] == "Manhattan"]
bk = mobility[mobility['Area'] == "Brooklyn"]
bx = mobility[mobility['Area'] == "Bronx"]
qns = mobility[mobility['Area'] == "Queens"]
si = mobility[mobility['Area'] == "Staten Island"]

```

```python
import altair as alt

source = nyc

chart = alt.Chart(source).mark_line().encode(
    x=alt.X('Date', title="Month"),
    y=alt.Y('Change compared to baseline', axis=alt.Axis(format='%'), scale=alt.Scale(domain=(-1, 1))),
    color='Mode of Transportation'
).properties(
    title='New York City, Total'
)

chart.configure_title(
    fontSize=20,
    font='Helvetica',
    anchor='middle',
    color='black'
)
```
