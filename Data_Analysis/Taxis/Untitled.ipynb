# Rideshares

```python
file_names = ['fhvhv_tripdata_2020-01.csv',
'fhvhv_tripdata_2020-02.csv',
'fhvhv_tripdata_2020-03.csv',
'fhvhv_tripdata_2020-04.csv',
'fhvhv_tripdata_2020-05.csv',
'fhvhv_tripdata_2020-06.csv',
'fhvhv_tripdata_2020-07.csv',
'fhvhv_tripdata_2020-08.csv',
'fhvhv_tripdata_2020-09.csv',
'fhvhv_tripdata_2020-10.csv',
'fhvhv_tripdata_2020-11.csv',
'fhvhv_tripdata_2020-12.csv']
```

```python
def agg_csv(file_name):
    df = pd.read_csv(file_name, usecols=['pickup_datetime', 'dropoff_datetime'])

    df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
    df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
    df['duration'] = df['dropoff_datetime'] - df['pickup_datetime']
    df['duration'] = df['duration'].dt.total_seconds()

    gb = df.groupby([df['pickup_datetime'].dt.month, df['pickup_datetime'].dt.day])

    agg_df = gb.count()
    agg_df['duration_mean'] = gb.duration.mean()
    agg_df['duration_median'] = gb.duration.median()

    agg_df['count'] = agg_df['pickup_datetime']
    del agg_df['pickup_datetime']
    del agg_df['dropoff_datetime']
    del agg_df['duration']
    
    return agg_df
```

```python
df_list = []

for file_name in file_names:
    agg = agg_csv(file_name)
    df_list.append(agg)
    print("hi")
    
```

```python
df_list
```

```python
rideshare = pd.concat(df_list, axis=0)
```

```python
rideshare.to_csv("rideshare_monthly_data.csv")
```
