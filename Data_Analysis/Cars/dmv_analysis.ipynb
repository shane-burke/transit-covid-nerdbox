# Monthly registration trends

```python
reg_date =  march21['Reg Valid Date'].value_counts().to_frame().reset_index().sort_values(by='index', ascending=True)
```

```python
reg_date['index'] = reg_date['index'].astype('str')
```

```python
reg_date['date'] = pd.to_datetime(reg_date['index'])
```

```python
reg_date_clipped = reg_date[reg_date['date'] > "2020-01-01"]
```

```python
reg_date_clipped['month'] = reg_date_clipped['date'].dt.month
```

```python
reg_date_clipped.to_csv("reg_dates.csv")
```

```python
import altair as alt
from vega_datasets import data

source = reg_date_clipped

alt.Chart(source).mark_line().encode(
    alt.X('date:T'),
    alt.Y('Reg Valid Date')
)
```

```python
source = reg_date_clipped

alt.Chart(source).mark_bar().encode(
    x='date',
    y='Reg Valid Date',
    # The highlight will be set on the result of a conditional statement
    color=alt.condition(
        alt.datum.year == 1810,  # If the year is 1810 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(width=600)
```

```python
reg_date_april =  april20['Reg Valid Date'].value_counts().to_frame().reset_index().sort_values(by='index', ascending=True)
```

```python
reg_date_april['index'] = reg_date_april['index'].astype('str')
reg_date_april['date'] = pd.to_datetime(reg_date_april['index'])
reg_date_april_clipped = reg_date_april[reg_date_april['date'] > "2018-04-01"]
reg_date_april_clipped['month'] = reg_date_april_clipped['date'].dt.month
```
